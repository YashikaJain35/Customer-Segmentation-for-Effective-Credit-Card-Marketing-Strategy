```python
from sklearn.preprocessing import StandardScaler
```

```python
X = pd.DataFrame(StandardScaler().fit_transform(df))
```

```python
import yellowbrick
import scipy.cluster.hierarchy as shc
import matplotlib.patches as patches
from matplotlib.patches import Rectangle
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import davies_bouldin_score, silhouette_score, calinski_harabasz_score
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer
from yellowbrick.style import set_palette
```

```python
# --- Transform into Array ---
X = np.asarray(X)

# --- Applying PCA ---
pca = PCA(n_components=2, random_state=24)
X = pca.fit_transform(X)
```

```python
elbow_score = KElbowVisualizer(KMeans(random_state=32, max_iter=500), k=(2, 10))
elbow_score.fit(X)
elbow_score.finalize()
elbow_score.show()
plt.show()
```

```python
elbow_score_ch = KElbowVisualizer(KMeans(random_state=32, max_iter=500), k=(2, 10), metric='calinski_harabasz', timings=False)
elbow_score_ch.fit(X)
elbow_score_ch.finalize()
elbow_score.show()
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=32, max_iter=500)
y_kmeans = kmeans.fit_predict(X)
```

```python
labels = kmeans.predict(X)
```

```python
centroids = kmeans.cluster_centers_
```

```python
labels
```

```python
centroids
```

```python
labels = kmeans.labels_

# Calculate the Davies-Bouldin Index
db = davies_bouldin_score(X, labels)

# Calculate the Silhouette Score
silhouette = silhouette_score(X, labels)

# Calculate the Calinski-Harabasz Index
ch = calinski_harabasz_score(X, labels)

# Print the results
print("Davies-Bouldin Index:", db)
print("Silhouette Score:", silhouette)
print("Calinski-Harabasz Index:", ch)
```

```python
cluster_colors=['#FFBB00', '#3C096C', '#9D4EDD', '#FFE270']
labels = ['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Centroids']
```

```python
s_viz = SilhouetteVisualizer(kmeans, colors=cluster_colors)
s_viz.fit(X)
s_viz.finalize()
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
# Fit the data to the model
dbscan.fit(X)

# Predict the clusters for the data points
labels = dbscan.labels_

# Print the cluster labels
print(labels)
```

```python
labels = dbscan.labels_

# Calculate the Davies-Bouldin Index
db = davies_bouldin_score(X, labels)

# Calculate the Silhouette Score
silhouette = silhouette_score(X, labels)

# Calculate the Calinski-Harabasz Index
ch = calinski_harabasz_score(X, labels)

# Print the results
print("Davies-Bouldin Index:", db)
print("Silhouette Score:", silhouette)
print("Calinski-Harabasz Index:", ch)
```
